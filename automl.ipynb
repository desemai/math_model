```python
from autogluon.tabular import TabularDataset, TabularPredictor
import pandas as pd
train_df = pd.read_csv("data_convert/train.csv", encoding='utf-8')
test_df = pd.read_csv("data_convert/test.csv", encoding='utf-8')
```

```python
label = 'Core_loss'
save_path = "agModels_Regression"
train_data = TabularDataset(train_df)
test_data = TabularDataset(test_df)
```

```python
predictor = TabularPredictor(label=label, path=save_path, verbosity=2, problem_type='regression')
predictor.fit(train_data, time_limit=2 * 3600, presets='best_quality', num_gpus=1, num_bag_folds=6, num_bag_sets=1, num_stack_levels=1)
```

```python
predictor.model_best
```

```python
y_pred = predictor.predict(test_data.drop(columns=[label]))
y_pred.head()
```

```python
predictor.evaluate(test_data, silent=True)
```

```python
predictor.leaderboard(test_data)
```

```python
submit_df = pd.read_csv("data_convert/submit.csv", encoding='utf-8')
submit_data = TabularDataset(submit_df)
predictor = TabularPredictor.load(save_path)
y_pred = predictor.predict(submit_data)
y_pred.to_csv("src/submit.csv", encoding='utf-8', index=True)
y_pred
```
